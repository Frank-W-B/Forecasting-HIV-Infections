```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
hiv_df = pd.read_csv("OpioidDB_HIVincidence_countydata_2017-12-21.csv")
hiv_df.head()
```

```python
hiv_df['county_code'] = hiv_df.STATEFP*1000 + hiv_df.COUNTYFP
hiv_df.head()
```

```python
msm_df = pd.read_csv("US MSM Estimates Data 2013.csv")
msm_df['county_code'] = msm_df.STATEFP*1000 + msm_df.COUNTYFP
msm_df.head()
```

```python
hiv_df['VALUE'].fillna(0, inplace=True)
hiv_df.head()
```

```python
plt.hist(hiv_df.VALUE, bins=1000)
plt.xlim(xmin=0, xmax=100)
plt.ylim(ymin=0, ymax=700)
plt.title('HIV incidence by county')
```

```python
hiv_df.describe()
```

```python
plt.hist(msm_df.MSM12MTH, bins=10000)
plt.xlim(xmin=0, xmax=2000)
plt.ylim(ymin=0, ymax=500)
plt.title('Number of MSM by county')
```

```python
msm_df.describe()
```

```python
msm_hiv_df = msm_df.merge(hiv_df, on='county_code', )
msm_hiv_df.head()
```

```python
plt.scatter(msm_hiv_df.MSM12MTH, msm_hiv_df.VALUE, )
plt.xlim(xmin=0, xmax=100000)
plt.ylim(ymin=0, ymax=300)
plt.title('Number of MSM vs HIV incidence')
plt.xlabel('Number of MSM')
plt.xlabel('HIV incidence')
```
